```python
import requests 
import os
from PIL import Image
from IPython.display import IFrame
```

2. <b>from PIL import Image:</b> This line imports the Image class from the PIL (Python Imaging Library) module. This library enables working with various image file formats (like JPEG, PNG, etc.) and manipulating images.

```python
url = 'https://s3.amazonaws.com/coursera_assets/meta_images/generated/CERTIFICATE_LANDING_PAGE/CERTIFICATE_LANDING_PAGE~LNZE8LTSH9ZE/CERTIFICATE_LANDING_PAGE~LNZE8LTSH9ZE.jpeg'
```

You can make a <code>GET</code> request via the method <code>get</code> to [www.coursera.com](https://s3.amazonaws.com/coursera_assets/meta_images/generated/CERTIFICATE_LANDING_PAGE/CERTIFICATE_LANDING_PAGE~LNZE8LTSH9ZE/CERTIFICATE_LANDING_PAGE~LNZE8LTSH9ZE.jpeg): to view my professional certificate from IBM through Coursera


```python
r= requests.get(url)
```

We have the response object <code>r</code>, this has information about the request, like the status of the request. We can view the status code using the attribute <code>status_code</code>.


```python
r.status_code
```

You can view the request headers:


```python
r.request.headers
```

You can view the <code>HTTP</code> response header using the attribute <code>headers</code>. This returns a python dictionary of <code>HTTP</code> response headers.


```python
r.headers
```

<code>Content-Type</code> indicates the type of data: this allows us save the file effectively

```python
r.headers['content-type']
```

An image is a response object that contains the image as a bytes-like object</a>. As a result, we must save it using a file object. First, we specify the <u>file path and
name</u>


```python
path = os.path.join(os.getcwd(), 'my_cert.jpeg')
```

```python
#we save the file using the `with` statement and accesing the content of the response we use the 'content'
with open(path, 'wb')as f:
    f.write(r.content)
```

We can view the image: using the image Library `Image`

```python
Image.open(path)
```
